```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
from scipy.stats import linregress
```

```python
data_df = pd.read_csv("crime_per_capita_housing_df.csv")
```

```python
data_df.head()
```

```python
# Build scatter plot for crime vs. median housing price
plt.figure(figsize=(8,5))
plt.scatter(x=data_df["2016 Crime per Capita (%)"], y=data_df["2016 Median Housing"], marker="o", facecolors="blue", edgecolors="black")
plt.xlabel("Crimes Per Capita (%)")
plt.ylabel("Median House Price ($)")
plt.title("Crime vs. Median House Price in California Counties in 2016")
plt.grid()
x_values = data_df.loc[:,"2016 Crime per Capita (%)"]
y_values = data_df.loc[:,"2016 Median Housing"]

#Find slope, y-intercept, r, pvalue, and standard deviation.  Put these values into a linear regression
(slope, intercept, rvalue, pvalue, stdev) = linregress(x_values, y_values)

#Develop linear regression
line_eq = "y = " + str(round(slope, 2)) + "x + " + str(round(intercept,2))
regress_values = x_values * slope + intercept

#Plot linear regression and rvalue
plt.plot(x_values, regress_values , "r-")
plt.annotate(line_eq, (4,500000), fontsize=12, color="red")
plt.annotate(f"r = {round(rvalue,6)}", (5.05,210000), color="teal")
print(line_eq)

#print r-value
print(f"The r-value is {rvalue}")

# Save the figure
plt.savefig("crime_housing_scatter/CrimevHouse16.png")

# Show plot
plt.show()
```
