```python
import pandas as pd
import matplotlib as plt
```

```python
#read csv and print dataframe
csv = "US Population by County.csv"
uspop_data = pd.read_csv(csv, engine='python', encoding='latin1')
uspop_data.head()
```

```python
#remove miscellaneous columns
califpop_df = pd.DataFrame(uspop_data.loc[uspop_data["STNAME"]=="California"])
califpop_df = califpop_df.drop(columns = ["STNAME"])
califpop_df.reset_index()
califpop_df=califpop_df.dropna(how='any')
califpop_df
```

```python
#print dataframe into csv file
califpop_df.to_csv('califpop_df.csv')
```
