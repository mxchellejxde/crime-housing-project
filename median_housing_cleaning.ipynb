```python
#Import dependencies

import matplotlib.pyplot as plt
import glob
import numpy as np
import pandas as pd
import os
```

```python
#Determine location of files to append into one file

path = os.getcwd()
print(path)
folder_path = "\Median Housing/*.csv"
folder = path+folder_path
all_files = glob.glob(os.path.join(folder))
housing_by_county = []

for file in all_files:
    housing_data = pd.read_csv(file,encoding="utf-8-sig")
    housing_df = pd.DataFrame(housing_data.iloc[9])
    housing_df['Year'] = os.path.basename(file)
    housing_by_county.append(housing_df)
    
housing_by_county
```

```python
#combine all the median housing by county lists into one dataframe
all_housing_df = pd.concat(housing_by_county)
all_housing_df.reset_index(inplace=True)

#remove unnecessary string characters
all_housing_df = all_housing_df.replace(', California!!Owner-occupied housing units with a mortgage!!Estimate','', regex=True)

#remove percent owner-occupied
all_housing_df = all_housing_df[all_housing_df['index'].str.contains('Percent')==False]
all_housing_df = all_housing_df[all_housing_df['index'].str.contains('Label')==False]

#remove the last 4 characters (" Year.csv") from the year column
all_housing_df['Year'] = all_housing_df['Year'].astype(str).str[:-4].astype(str)

#remove commas from string and convert to float
all_housing_df[[9]] = all_housing_df[[9]] .replace(',', '', regex=True).astype(float)

#pivot
housing_df = all_housing_df.pivot(columns='Year',index='index', values=9)

housing_df.head()
```

```python
#print dataframe into csv file
housing_df.to_csv('housing_df.csv')
```
